```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([0.2,2,3,4])
for  i in range(3):
    x[i+1]=4*x[i]*(1-x[i])
print(x)
```

```python
y=np.empty_like(x)
mu=0.001
for i in range(3):
    y[i]=x[i]+np.random.normal(0,mu)
plt.plot(y,'b*',markersize=4)
plt.xlabel('y')
plt.ylabel('x')
plt.show()
```

```python
#np.random.seed(42)
def j(x2,y2,y3):
    return (x2-y2)**2+(4*x2**2-4*x2**2-y3)**2
x2=np.random.uniform(-0.1,1.1,4)
y2=y[1]
y3=y[2]
X=j(x2,y2,y3)
```

```python
plt.plot(X,x2)
plt.xlabel('j')
plt.ylabel('x2')
plt.show()
```
